### CNN - Sequential  Model

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, BatchNormalization, Dense, Flatten
```

```python
cnn_smodel= Sequential()
```

### First Convolutional Layer

```python
cnn_smodel.add( Conv2D(32, (3,3), activation  ="relu", strides = (1,1), 
                padding = "valid", input_shape = (48,48,3))  )


cnn_smodel.add( MaxPooling2D((2,2), strides = 2, padding = "valid"))

cnn_smodel.add(Dropout(0.20))

cnn_smodel.add(BatchNormalization(axis =-1))
```

### Second Convolutional Layer

```python
cnn_smodel.add( Conv2D(16, (3,3), activation  ="relu", strides = (1,1), padding = "valid")  )

cnn_smodel.add( MaxPooling2D((2,2), strides = 2, padding = "valid"))

cnn_smodel.add(Dropout(0.20))

cnn_smodel.add(BatchNormalization(axis =-1))
```

### Third Convolutional Layer

```python
cnn_smodel.add( Conv2D(32, (3,3), activation  ="relu", strides = (1,1), padding = "valid")  )

cnn_smodel.add( MaxPooling2D((2,2), strides = 2, padding = "valid"))

cnn_smodel.add(Dropout(0.20))

cnn_smodel.add(BatchNormalization(axis =-1))
```

### Fully Connected Layer  - ANN

```python
cnn_smodel.add(Flatten())
```

```python
cnn_smodel.add(Dense(64, activation = "relu"))
```

```python
cnn_smodel.add(Dense(32, activation = "relu"))
```

```python
cnn_smodel.add(Dense(16, activation = "relu"))
```

```python
cnn_smodel.add(Dense(32, activation = "relu"))
```

```python
cnn_smodel.add(Dense(64, activation = "relu"))
```

```python
cnn_smodel.add(Dense(7, activation = "softmax"))
```

```python
cnn_smodel.summary()
```

```python
cnn_smodel.compile(
                    
                    optimizer="adam",
                    loss = "sparse_categorical_crossentropy",
                    metrics = ["accuracy"]

                    )
```

```python
cnn_smodel.fit(Xtrain, ytrain, epochs = 40)
```

```python
cnn_smodel.evaluate(Xtest,ytest)
```
